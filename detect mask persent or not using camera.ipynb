```python
from tensorflow.keras.models import load_model 
model = load_model("maskdetecter.h5")  
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
image = np.array(plt.imread('machine_learning/flowers/face.jpg').astype(float))
image = image/255.0
my_image = cv2.resize(image, dsize=(150,150)).reshape((1,150,150,3))
num = model.predict(my_image)
plt.imshow(image)
if num>0.5:
    print("I think mask is present on the face of man"+str(num))
elif num<0.5:
      print("I think mask is not present on the face of man"+str(num))
```

```python
import zipfile
import random
import shutil
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from shutil import copyfile
from os import getcwd
from os import listdir
from sklearn.metrics import f1_score
from sklearn.utils import shuffle
import matplotlib.image  as mpimg
```

```python
labels_dict={1:'with_mask',0:'without_mask'}
color_dict={0:(0,0,255),1:(0,255,0)}

size = 2
webcam = cv2.VideoCapture(0) #Use camera 0

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

while True:
    (rval, im) = webcam.read()
    im=cv2.flip(im,1,1) #Flip to act as a mirror

    # Resize the image to speed up detection
    mini = cv2.resize(im, (im.shape[1] // size, im.shape[0] // size))

    # detect MultiScale / faces 
    faces = face_cascade.detectMultiScale(mini)

    # Draw rectangles around each face
    for f in faces:
        (x, y, w, h) = [v * size for v in f] 
        face_img = im[y:y+h, x:x+w]
        image = face_img/255.0
        my_image = cv2.resize(image, dsize=(150,150)).reshape((1,150,150,3))
        num = model.predict(my_image)
        #print(num)
        if num>0.5:
            label=0
        elif num<0.5:
            label=1
        num=str(num)
        print(num)    
        cv2.rectangle(im,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(im,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(im,labels_dict[label]+str(num), (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
    # Show the image
    cv2.imshow('LIVE',   im)
    key = cv2.waitKey(10)
    if key == 27:
        break

webcam.release()

# Close all started windows
```
